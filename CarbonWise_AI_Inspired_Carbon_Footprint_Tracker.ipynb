```python
import gradio as gr
import random

def carbonwise(distance, transport, electricity, diet, shopping, plastic):

    transport_factors = {
        "Car": 0.21,
        "Bike": 0.09,
        "Bus": 0.05,
        "Train": 0.03,
        "Walk/Cycle": 0
    }

    diet_factors = {
        "Non-Veg": 3.3,
        "Vegetarian": 2.0,
        "Vegan": 1.5
    }

    shopping_factors = {
        "Low": 5,
        "Medium": 15,
        "High": 30
    }

    travel_emission = distance * transport_factors[transport] * 30
    electricity_emission = electricity * 0.82
    diet_emission = diet_factors[diet] * 30
    shopping_emission = shopping_factors[shopping]
    plastic_emission = plastic * 0.1 * 30

    total = travel_emission + electricity_emission + diet_emission + shopping_emission + plastic_emission

    if total < 120:
        category = "🟢 Low Carbon Footprint"
        message = "Great! Your lifestyle is already quite eco-friendly."
    elif total < 250:
        category = "🟡 Medium Carbon Footprint"
        message = "You are doing okay, but small changes can reduce your emissions."
    else:
        category = "🔴 High Carbon Footprint"
        message = "Your emissions are high. Focus on travel, electricity, and lifestyle changes."

    green_score = max(0, min(100, int(100 - total / 4)))

    suggestions = []

    if transport in ["Car", "Bike"]:
        suggestions.append("Use public transport, walking, or cycling for short-distance travel.")
    else:
        suggestions.append("Your transport choice is comparatively eco-friendly. Keep it up.")

    if electricity > 100:
        suggestions.append("Reduce electricity usage by switching off unused lights and using LED bulbs.")
    else:
        suggestions.append("Your electricity usage is under control.")

    if diet == "Non-Veg":
        suggestions.append("Try 1–2 vegetarian meals per week to reduce food-related emissions.")
    elif diet == "Vegetarian":
        suggestions.append("A vegetarian diet helps reduce carbon emissions compared to high meat consumption.")
    else:
        suggestions.append("A vegan diet has one of the lowest food-related carbon impacts.")

    if shopping == "High":
        suggestions.append("Avoid unnecessary online shopping and choose durable, reusable products.")
    elif shopping == "Medium":
        suggestions.append("Try combining orders and buying only what you really need.")
    else:
        suggestions.append("Low shopping frequency helps reduce packaging and delivery emissions.")

    if plastic > 3:
        suggestions.append("Carry a reusable bottle and cloth bag to reduce single-use plastic.")
    else:
        suggestions.append("Your plastic usage is low. Continue avoiding single-use plastic.")

    challenges = [
        "No Plastic Day: Avoid single-use plastic for one full day.",
        "Public Transport Week: Use bus/train/cycle for at least 3 days.",
        "Energy Saver Challenge: Switch off all unused appliances daily.",
        "Green Meal Challenge: Eat vegetarian food for 2 days this week.",
        "Reusable Bottle Challenge: Carry your own water bottle every day."
    ]

    selected_challenge = random.choice(challenges)

    weekly_reduction = total * 0.12
    yearly_reduction = weekly_reduction * 12

    report = f"""
# 🌱 CarbonWise Report

## Total Monthly Carbon Footprint
**{total:.2f} kg CO₂/month**

## Carbon Category
**{category}**

## Green Score
**{green_score}/100**

## Insight
{message}

---

## Emission Breakdown
- Travel Emission: **{travel_emission:.2f} kg CO₂/month**
- Electricity Emission: **{electricity_emission:.2f} kg CO₂/month**
- Diet Emission: **{diet_emission:.2f} kg CO₂/month**
- Shopping Emission: **{shopping_emission:.2f} kg CO₂/month**
- Plastic Emission: **{plastic_emission:.2f} kg CO₂/month**

---

## Personalized Action Plan
1. {suggestions[0]}
2. {suggestions[1]}
3. {suggestions[2]}
4. {suggestions[3]}
5. {suggestions[4]}

---

## Weekly Green Challenge
✅ **{selected_challenge}**

---

## Possible Impact
If you follow these actions, you may reduce around **{weekly_reduction:.2f} kg CO₂/month**.

Estimated yearly reduction: **{yearly_reduction:.2f} kg CO₂/year**

---

## Final Message
Small daily actions like saving electricity, reducing plastic, using public transport, and making mindful food choices can create a big positive impact on the environment.
"""

    return report


app = gr.Interface(
    fn=carbonwise,
    inputs=[
        gr.Number(label="Daily Travel Distance in KM", value=10),
        gr.Dropdown(["Car", "Bike", "Bus", "Train", "Walk/Cycle"], label="Transport Mode", value="Bike"),
        gr.Number(label="Monthly Electricity Units", value=120),
        gr.Dropdown(["Non-Veg", "Vegetarian", "Vegan"], label="Diet Type", value="Vegetarian"),
        gr.Dropdown(["Low", "Medium", "High"], label="Online Shopping Frequency", value="Medium"),
        gr.Number(label="Single-use Plastic Items Used Daily", value=2)
    ],
    outputs=gr.Markdown(label="CarbonWise Personalized Report"),
    title="🌱 CarbonWise - AI Inspired Carbon Footprint Tracker",
    description="""
CarbonWise helps individuals understand, track, and reduce their carbon footprint through simple lifestyle inputs, personalized insights, green score, and weekly eco-challenges.
"""
)

app.launch(share=True)
```
